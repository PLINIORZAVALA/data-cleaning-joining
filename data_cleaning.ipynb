```python
import pandas as pd
import os

# Configuración
base_path = 'lista_certificados'

# 1. Identificar un archivo de ejemplo
files = [f for f in os.listdir(base_path) if f.endswith('.xlsx')]
if files:
    sample_file = files[0]
    sample_file_path = os.path.join(base_path, sample_file)

    print(f"Archivo de ejemplo seleccionado: {sample_file}")

    # 2. Leer las hojas del archivo
    xls = pd.ExcelFile(sample_file_path)
    sheet_names = xls.sheet_names
    print(f"Hojas encontradas: {sheet_names}")

    # 3. Leer cada hoja individualmente
    # Se asume que hay dos hojas según tu descripción
    if len(sheet_names) >= 2:
        df_hoja1 = pd.read_excel(sample_file_path, sheet_name=0)
        df_hoja2 = pd.read_excel(sample_file_path, sheet_name=1)

        print("\n--- Vista previa Hoja 1 ---")
        display(df_hoja1.head())
        print(f"Dimensiones: {df_hoja1.shape}")

        print("\n--- Vista previa Hoja 2 ---")
        display(df_hoja2.head())
        print(f"Dimensiones: {df_hoja2.shape}")
    else:
        print("El archivo no tiene al menos 2 hojas.")
else:
    print("No se encontraron archivos .xlsx en la carpeta.")
```
